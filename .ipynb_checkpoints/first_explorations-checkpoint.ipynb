## Cleaning of the dataset

```python
# making everything lower case 
text_lower = text_woolf.lower()

# eliminating the special characters
# I am doing this to have a better look at the text and for later analysis purposes. "´s" etc. are not important for the analysis as the whole words are more importand and have more measning.

clean_text = re.sub("[^a-z']", ' ', text_lower)
clean_text

# transforming multiple spaces to one
clean_text = re.sub("[ ]+", ' ', clean_text)

#checking if it worked
clean_text
```

## 4. How big is Woolf´s vocabulary? (unique words) #####

```python
# this could be an interesting thing to know in regard to the complexity of the text for further analysis

wordfrequency_woolf = Counter(list_words)

print (f"There are {len(wordfrequency_woolf)} unique words in woolfs vocabulary in 'A Room of One´s Own'")
```

## 5. What are the 20 words (apart from "and", "the", etc.) that Virginia Woolf uses the most? #####

```python
wordfrequency_woolf.most_common(20)
```

### Creating a bar chart

```python
labels, values = zip(*wordfrequency_woolf.most_common(20))

indexes = np.arange(len(labels))
width = 0.5

plt.bar(indexes, values, width)
fig = plt.gcf()
fig.set_size_inches(18.5, 10.5)
plt.xticks(indexes + width * 0.5, labels)
plt.show()
```

## 6. Which words/sequences occur most frequently in consecutive combinations? #####

### Bigrams

```python
two_words = []
for i in range(len(list_words)-1):
    tw=list_words[i]+" "+list_words[i+1]
    two_words.append(tw)
two_words
count_bigram = {}
for t in two_words:
    if t in count_bigram:
        count_bigram[t] += 1
    else: 
        count_bigram[t] = 1
count_bigram_sorted = {k: v for k, v in sorted(count_bigram.items(), key=lambda item: item[1], reverse = True)}
most_frequent_bigram = list(count_bigram_sorted.keys())[0:10]
most_frequent_bigram
```
